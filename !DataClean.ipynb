```python
df = pd.read_csv('DATA.csv')

# Change to lower
df["Text"] = df["Text"].str.lower()

# only get that has the word 'jokowi'
df = df[df["Text"].str.contains("jokowi")]

df["Date"] = pd.to_datetime(df["Date"])

plotDay = []
for i in df["Date"].dt.date.unique():
    plotDay.append([i, len(df[df["Date"].dt.date == i])])
```

```python
plt.figure(figsize=(20, 10))
plt.bar([i[0] for i in plotDay], [i[1] for i in plotDay], color='b')
plt.xlabel('Date')
plt.ylabel('Number of Mentions')
plt.title('Number of Mentions of "Jokowi" Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
hourLen = []
for i in df["Date"].dt.hour.unique():
    hourLen.append([i, len(df[df["Date"].dt.hour == i])])
```

```python
sorted(hourLen, key=lambda x: x[0])
sorted_hourLen = sorted(hourLen, key=lambda x: x[0])

# Plotting the sorted hourLen data
hours = [i[0] for i in sorted_hourLen]
counts = [i[1] for i in sorted_hourLen]

plt.figure(figsize=(20, 10))
plt.bar(hours, counts, color='g')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Mentions')
plt.title('Number of Mentions of "Jokowi" by Hour of the Day 2023-2024')
plt.xticks(hours)
plt.grid(True)
plt.show()
```
